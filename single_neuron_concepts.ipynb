# Single Neuron Concepts:

- Neurons are the basic units of a neural network.
- A single neuron receives multiple inputs and creates a single output.
- Number of inputs to a neuron is detrmined by the dimensionality of the input space.
- Neurons have associated weights for each input, indicating the strength of the connection.
- The $net$ value of a node is calculated as $\mathbf{net=\Sigma_{i=1}^{n} (w_i*x_i)+b}$

  - $x_i$ represent the $i_{th}$ input and $w_i$ is the weight associated with the $i_{th}$ input.
  - $b$ is a constant called bias.
  - $n$ is the number of inputs to a neuron.
- The equation $\mathbf{net=\Sigma_{i=1}^{n} (w_i*x_i)+b}$ can be interpreted (presented) as a hyper-plane in $n+1$ dimensional space.
- The intersection of $\mathbf{net=\Sigma_{i=1}^{n} (w_i*x_i)+b}$ , which is an $n+1$-dimensional hyper-plane, with the input space creates an $n$-dimensional hyper-plane. This $n$-dimensional hyper-plane is called decision boundary.
- Decision boundary separates input space into two different regions.

- The output of a neuron is calculated by passing the $net$ value through a function which is usually called an activation function or transfer function.
- The activation function is introduced to create a non-linearity in the output of a neuron.
- Common activation functions include sigmoid, tanh, and rectified linear unit (ReLU).
- Single neurons have limitations in handling complex relationships within data.
- A multi-layer neural networks is made of multiple neurons which are arranged in multiple layers of neurons.
